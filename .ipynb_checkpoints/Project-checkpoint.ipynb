### Automatically generating appropriate visualizations for every feature in the dataset. Selecting bar charts for categorical or discrete features and histograms for continuous numeric variables. This allows quick and effective inspection of variable distributions, detection of imbalances, and identification of potential data issues.

```python
def displayGraphs(data):
    for col in data.columns:
        plt.figure()
    
        if data[col].dtype == "object":
            counts = data[col].value_counts().sort_index()  
            plt.bar(counts.index, counts.values)
            plt.xlabel(col)
            plt.ylabel("Frequency")
    
        elif data[col].dtype == "int64":
    
            if data[col].nunique() < 20:
                counts = data[col].value_counts().sort_index() 
                plt.bar(counts.index, counts.values)
                plt.xticks(counts.index)
                plt.ylabel("Frequency")
            else:
                plt.hist(data[col], bins=20)
                plt.ylabel("Count")
    
            plt.xlabel(col)
    
        elif data[col].dtype == "float64":
            plt.hist(data[col], bins=20)
            plt.xlabel(col)
            plt.ylabel("Count")
    
        plt.title(col)
        plt.tight_layout()
        plt.show()
```

```python
displayGraphs(data)
```

### Automatically generating plots for every feature with respect to the label

```python
def plot_features_with_label(data, label):
    for col in data.columns:
        if col == label:
            continue
        
        plt.figure(figsize=(7, 4))
        plt.title(f"{col} vs {label}")

        if data[col].dtype == "object" or data[col].nunique() < 10:
            sns.countplot(x=col, hue=label, data=data)
            plt.xlabel(col)
            plt.ylabel("Count")

        else:
            sns.boxplot(x=label, y=col, data=data)
            plt.xlabel(label)
            plt.ylabel(col)

        plt.tight_layout()
        plt.show()
```

```python
plot_features_with_label(data, "divorced")
```

### Checking the correlation between each feature and the label

```python
label_col = "divorced" 

numeric_cols = data.select_dtypes(include=["int64", "float64"]).columns.drop(label_col)

correlations = data[numeric_cols].corrwith(data[label_col])
print(correlations)

categorical_cols = data.select_dtypes(include=["object"]).columns

encoded = data[categorical_cols].apply(lambda x: x.astype('category').cat.codes)

cat_correlations = encoded.corrwith(data[label_col])
print(cat_correlations)
```

```python
all_correlations = pd.concat([correlations, cat_correlations])
all_correlations.sort_values().plot(kind="bar", figsize=(10,6))
plt.ylabel("Correlation with " + label_col)
plt.title("Feature correlations with label")
plt.show()
```

```python
numeric_cols = data.select_dtypes(include=['int64', 'float64'])

corr_matrix = numeric_cols.corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Feature Correlation Matrix")
plt.show()
```

The correlations between individual features and divorce are generally very weak, indicating that no single feature has a strong linear relationship with the outcome. The strongest positive associations are with domestic_violence_history, infidelity_occurred, and financial_stress_level, suggesting these factors slightly increase divorce risk. Weak negative correlations are observed for communication_score and trust_score, implying that better communication and higher trust are modestly protective. Most demographic and categorical variables show negligible linear correlation.

# Analyzing the second dataset

## The second dataset
The second dataset contains data about 170 couples with their corresponding Divorce Predictors Scale variables (DPS) on the basis of Gottman couples therapy.
The couples are from various regions of Turkey wherein the records were acquired from face-to-face interviews from couples who were already divorced or happily married.
All responses were collected on a 5 point scale (0=Never, 1=Seldom, 2=Averagely, 3=Frequently, 4=Always).

### Features:
A lot of the questions are ambiguous or very similar. We drop obscure and ambiguous questions and combine almost identical questions,
so only 1-3 questions from each category are used.

Strikedowned question => Question that we drop

Questions that have the symbol * in the ened => Questions that we combine (for each category combination 1: *, combination 2: ** etc)

### Reading in the second dataset

```python
data_turkish = pd.read_csv("divorce_data.csv", sep =";")
```

### Looking at the data

```python
data_turkish.info()
```

### Checking if all the feature values are between 0 and 4.

```python
cols = list(data_turkish.columns)
cols.remove('Divorce')

correct = "All values are correct"
for column in cols:
    if (data_turkish[column].between(0, 4).all() == False):
        correct = "Values are not correct!"
        
print(correct)
```

### Dropping confusing features, combining the features that are similar

```python
data_turkish = data_turkish.drop(columns = ["Q1", "Q3", "Q4", "Q11", "Q35", "Q36", "Q38", "Q39", "Q40", "Q42"])
FeaturesToCombine = [
    ["Q5", "Q8", "Q9"],
    ["Q6", "Q7"],
    ["Q10", "Q12", "Q13", "Q14", "Q15", "Q16", "Q17", "Q18", "Q19", "Q20"],
    ["Q21", "Q22", "Q23", "Q24", "Q25", "Q26", "Q27", "Q28", "Q29", "Q30"],
    ["Q32", "Q33", "Q34"],
    ["Q31", "Q37", "Q41"],
    ["Q43", "Q44", "Q45", "Q46", "Q47"],
    ["Q48", "Q49", "Q50", "Q51"],
    ["Q52", "Q53", "Q54"]
]

Titles = ["Spending time together",
                 "Not spending time together at home",
                 "Common values and similarities",
                 "Knowledge about spouse",
                 "Usage of hostile and negative statements",
                 "Feeling intense negative emotions in disagreements",
                 "Handling conflict by remaining silent or withdrawing",
                 "Often feeling that their partner is to blame",
                 "Frequently pointing out their partner’s shortcomings"
]

# Rename the Q2
data_turkish.rename(columns={'Q2': 'Setting aside differences if necessary'}, inplace=True)

# Create new features, drop old
for i in range(len(FeaturesToCombine)):
    data_turkish[Titles[i]] = data_turkish[FeaturesToCombine[i]].mean(axis = 1) 
    data_turkish = data_turkish.drop(columns = FeaturesToCombine[i])

# Reorder the Label so that it is last:
cols = list(data_turkish.columns)
cols.remove("Divorce")
cols.append("Divorce")
data_turkish = data_turkish[cols]
```
