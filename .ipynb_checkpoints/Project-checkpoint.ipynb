```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("divorce_df.csv")
```

```python
print(data.info())
data.head()
```

```python
print(data["communication_score"].between(0, 10).all())
print(data["financial_stress_level"].between(0, 10).all())
print(data["social_support"].between(0, 10).all())
print(data["trust_score"].between(0, 10).all())
```

```python
for col in data.columns:
    plt.figure()

    if data[col].dtype == "object":
        counts = data[col].value_counts().sort_index()  
        plt.bar(counts.index, counts.values)
        plt.xlabel(col)
        plt.ylabel("Frequency")

    elif data[col].dtype == "int64":

        if data[col].nunique() < 20:
            counts = data[col].value_counts().sort_index() 
            plt.bar(counts.index, counts.values)
            plt.xticks(counts.index)
            plt.ylabel("Frequency")
        else:
            plt.hist(data[col], bins=20)
            plt.ylabel("Count")

        plt.xlabel(col)

    elif data[col].dtype == "float64":
        plt.hist(data[col], bins=20)
        plt.xlabel(col)
        plt.ylabel("Count")

    plt.title(col)
    plt.tight_layout()
    plt.show()
```

```python
label_col = "divorced" 

numeric_cols = data.select_dtypes(include=["int64", "float64"]).columns.drop(label_col)

correlations = data[numeric_cols].corrwith(data[label_col])
print(correlations)

categorical_cols = data.select_dtypes(include=["object"]).columns

encoded = data[categorical_cols].apply(lambda x: x.astype('category').cat.codes)

cat_correlations = encoded.corrwith(data[label_col])
print(cat_correlations)

```

```python
import matplotlib.pyplot as plt

all_correlations = pd.concat([correlations, cat_correlations])
all_correlations.sort_values().plot(kind="bar", figsize=(10,6))
plt.ylabel("Correlation with " + label_col)
plt.title("Feature correlations with label")
plt.show()
```
